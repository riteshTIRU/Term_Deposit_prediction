```python
output = data['subscribed']
```

```python
data.drop(columns = ['subscribed'],inplace=True)
```

```python
lis = []
columns_names = data.columns
for i in columns_names:
  if(data[i].dtype == 'O'):
    lis.append(i)
```

```python
for i in lis:
  if(data[i].nunique()==2):
    data[i] = (data[i]=='yes').astype(int)
```

```python
lis = []
columns_names = data.columns
for i in columns_names:
  if(data[i].dtype == 'O'):
    lis.append(i)
```

```python
lis
```

```python
data = pd.get_dummies(data=data,columns=lis)
```

```python
data.info()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(data,output,test_size = 0.2,random_state=48 )
```

```python
from imblearn.over_sampling import SMOTE
os = SMOTE(random_state=48)
```

```python
os_X_train,os_y_train = os.fit_resample(X_train,y_train)
```

```python
os_y_train.value_counts()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lgr = LogisticRegression()
lgr.fit(os_X_train,os_y_train)
```

```python
pred = lgr.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print(accuracy_score(y_test,pred))
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc = DecisionTreeClassifier()

dtc.fit(os_X_train,os_y_train)
```

```python
pred = dtc.predict(X_test)
```

```python
print(accuracy_score(y_test,pred))
```

```python
from sklearn.metrics import confusion_matrix

print(confusion_matrix(y_test,pred))
```

```python
from sklearn.svm import SVC
model = SVC()
```

```python
model.fit(os_X_train,os_y_train)
pred = model.predict(X_test)
```

```python
print(accuracy_score(y_test,pred))
```

```python
print(confusion_matrix(y_test,pred))
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=2)
```

```python
knn.fit(os_X_train,os_y_train)
pred = knn.predict(X_test)
```

```python
print(accuracy_score(y_test,pred))
```

```python
print(confusion_matrix(y_test,pred))
```
